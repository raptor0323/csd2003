```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

url = "https://www.amazon.in"
req = requests.get(url)

print(req)
d = BeautifulSoup(req.text,"lxml")
print(d)

d1 = pd.DataFrame(d)
d1
d1.to_csv("AmazonData.csv")
```
